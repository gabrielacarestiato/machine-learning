# 1. Escopo, objetivo e definição do problema




### 1.1 Problema

O futebol é um esporte em que o desempenho dos atletas depende fortemente de sua condição física. Lesões podem comprometer não apenas a carreira dos jogadores, mas também o rendimento de seus times ao longo da temporada. Nesse contexto, compreender os padrões dessas lesões e seus efeitos no retorno dos atletas ao campo é fundamental para embasar decisões técnicas, médicas e gerenciais.

O conjunto de dados utilizado neste projeto é o Player Injuries and Team Performance Dataset, disponível no Kaggle, que reúne informações sobre lesões de jogadores de oito clubes da Premier League entre 2019 e 2023, bem como indicadores de desempenho de seus times.

O objetivo do projeto é investigar padrões de lesões e seus impactos sobre o desempenho dos jogadores e equipes ao longo das temporadas. Além da análise exploratória, o estudo aplica modelos de classificação supervisionada para prever a classe de duração da lesão, oferecendo insights relevantes para clubes e comissões técnicas no planejamento de escalações e estratégias.

### 1.2 Tipo de Tarefa

O presente trabalho trata de um problema de classificação supervisionada, uma vez que a variável-alvo é categórica e o modelo é treinado a partir de dados rotulados. O objetivo é prever a classe de duração da lesão (curta ou longa), com base em características dos jogadores lesionados, como idade, posição e tipo de lesão.

A variável-alvo, classe_duração, foi criada a partir da categorização da duração em dias (diferença entre a data de retorno e a data da lesão). Assim, em vez de prever diretamente o número de dias, o modelo aprende a classificar o jogador em uma das categorias pré-definidas.

Essa abordagem é útil para identificar rapidamente o grau esperado de afastamento, permitindo apoiar o planejamento médico, tático e de escalação das equipes, sem necessidade de estimar valores numéricos exatos.

### 1.3 Área de Aplicação

O projeto se insere no campo de análise de dados no esporte, com foco em saúde e performance de atletas. A análise preditiva contribui tanto para a prevenção de lesões quanto para o suporte ao diagnóstico médico, auxiliando equipes médicas e técnicas na tomada de decisão. Além disso, gera benefícios diretos para a gestão dos clubes, permitindo um planejamento mais eficiente de elenco, treinamentos e estratégias de escalação, o que pode resultar em melhor desempenho esportivo ao longo da temporada.

Do ponto de vista técnico, o projeto se baseia em dados tabulares, pois utiliza informações estruturadas sobre jogadores, lesões e desempenho dos times. As variáveis numéricas e categóricas, como idade, posição, tipo de lesão e duração do afastamento, são organizadas em formato de tabela e servem de insumo para a construção de modelos de classificação supervisionada.

 ### 1.4 Valor para o negócio/usuário

 O projeto gera valor ao transformar dados de lesões em insights práticos para diferentes públicos no futebol. Para as equipes médicas, fornece estimativas confiáveis da classe de duração da lesão e identificação de fatores que prolongam afastamentos. ### 2.2 Funções python
Foram definidas funções específicas para avaliar modelos de classificação. Cada função retorna um dicionário com as principais métricas de desempenho, permitindo comparar diferentes algoritmos de forma padronizada. Essa abordagem reduz repetição de código, melhora a legibilidade e garante consistência na avaliação dos experimentos.

```python
def evaluate_classification(y_true, y_pred, proba=None):
    acc = accuracy_score(y_true, y_pred)
    f1w = f1_score(y_true, y_pred, average="weighted")
    auc = roc_auc_score(y_true, proba[:,1]) if (proba is not None and proba.shape[1]==2) else np.nan
    return {"accuracy": acc, "f1_weighted": f1w, "roc_auc": auc}
```

```python
#Cálculo do IQR
q1 = df['idade'].quantile(0.25)
q3 = df['idade'].quantile(0.75)
iqr = q3 - q1

#Limites para detectar outliers
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

#Filtragem dos outliers
outliers_inferiores = df[df['idade'] < limite_inferior]
outliers_superiores = df[df['idade'] > limite_superior]

#Exibição dos resultados
print("Outliers inferiores (idade < {:.2f}):".format(limite_inferior))
print(outliers_inferiores)

print("\nOutliers superiores (idade > {:.2f}):".format(limite_superior))
print(outliers_superiores)
```

Outliers inferiores: Nenhum jogador tem idade menor que 16,5 anos. Faz sentido, pois dificilmente atletas profissionais estão abaixo dessa idade.

Outliers superiores: Foram encontrados jogadores com 38 e 39 anos, como Cristiano Ronaldo e Ashley Young. Também faz sentido, já que são casos raros de longevidade na carreira.

Ao converter as datas, há valores ausentes. Isso acontece porque as datas estão despadronizadas, atrapalhando os cálculos de duração (dias), por exemplo:

Meses abreviados misturados com completos: o mês de janeiro aparece como Jan e January, para os meses de junho e julho isso também ocorre.
"Present" ao invés da data.
Datas escritas com vírgula colada no ano (ex: Sep 16,2020 em vez de Sep 16, 2020).
Erro no ano (ex: Mar 7, 0202 em vez de Mar 7, 2020).
Como essa variável faz parte da variável-alvo do problema, e a ausência de valor inviabiliza seu uso, serão removidos os registros "Present" (lesões ainda em andamento) e padronizadas as outras datas.

```python
#Criando novamente a cópia do dataset original para desfazer a conversão das datas e tratar os problemas
df_tratado = df.copy()

#Remoção das linhas com 'present' na coluna de retorno
df_tratado = df_tratado[~df_tratado['dt_retorno'].astype(str).str.lower().str.contains('present', na=False)]

#Correção das datas com vírgula colada no ano
df_tratado['dt_lesão'] = df_tratado['dt_lesão'].astype(str).str.replace(r',(?=\d)', ', ', regex=True)
df_tratado['dt_retorno'] = df_tratado['dt_retorno'].astype(str).str.replace(r',(?=\d)', ', ', regex=True)

#Correção de erro no ano
df_tratado['dt_lesão'] = df_tratado['dt_lesão'].str.replace('0202', '2020')
df_tratado['dt_retorno'] = df_tratado['dt_retorno'].str.replace('0202', '2020')

#Padronização dos formatos de datas (meses completos e abreviados misturados)
def padronizar_data(data_str):
    try:
        return parser.parse(data_str).strftime('%b %d, %Y')
    except:
        return None  #Retorna None se a conversão falhar

df_tratado['dt_lesão'] = df_tratado['dt_lesão'].apply(padronizar_data)
df_tratado['dt_retorno'] = df_tratado['dt_retorno'].apply(padronizar_data)

#Conversão das colunas de data
df_tratado['dt_lesão'] = pd.to_datetime(df_tratado['dt_lesão'], errors='coerce')
df_tratado['dt_retorno'] = pd.to_datetime(df_tratado['dt_retorno'], errors='coerce')

#Função para categorizar a lesão
def categorizar_lesao(dias):
    if dias <= 30:
        return "curta"
    else:
        return "longa"

#Criação da coluna de classe (sem manter a duração no df_tratado)
df_tratado['classe_duração'] = (
    (df_tratado['dt_retorno'] - df_tratado['dt_lesão']).dt.days
).apply(categorizar_lesao)

#Seleção das colunas finais para o modelo
df_tratado = df_tratado[['posição', 'idade', 'temporada', 'fifa', 'lesão', 'classe_duração']].copy()

#Visualizar
df_tratado.head(10)
```

Antes da conversão das colunas de data, foi realizada uma etapa de limpeza e padronização dos valores. Linhas contendo o termo “Present” foram removidas, pois impossibilitam o cálculo da duração da lesão. Além disso, foram corrigidas datas com vírgulas coladas ao ano, possíveis erros de digitação no ano e padronizados os nomes dos meses. Essas medidas garantiram que a conversão para o tipo datetime fosse realizada de forma segura e que a variável duração (dias) fosse corretamente calculada.

Como a classe de duração é a variável-alvo do problema de classificação supervisionada, a ausência ou inconsistência de valores inviabilizaria o uso desses registros no modelo. Por isso, as correções e exclusões foram fundamentais para preservar a qualidade dos dados.

## 4.3 Separação em conjuntos de treino e teste
Após a definição das variáveis independentes (features) e da variável-alvo (target), foi realizada a divisão dos dados em conjuntos de treino e teste.

Features: posição, idade, temporada, fifa e lesão.

Target: classe_duração (curta ou longa).

Tamanho dos conjuntos: 80% para treino e 20% para teste.

Estratificação: utilizada a opção stratify=y no train_test_split, garantindo que a proporção entre as classes seja mantida tanto no conjunto de treino quanto no de teste.

Semente global: random_state=SEED (42), assegurando reprodutibilidade da divisão.

```python
#Remover linhas com NaN no target
df_tratado = df_tratado.dropna(subset=["classe_duração"]).reset_index(drop=True)

#Definição de features e target
features = ['posição', 'idade', 'temporada', 'fifa', 'lesão']
target = 'classe_duração'

X = df_tratado[features].copy()
y = df_tratado[target].copy()

print("Target:", target)
print("N features:", len(features))
print("Total de registros após remoção de NaN:", df_tratado.shape[0])

#Divisão treino/teste para classificação
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=SEED, stratify=y
)

print("Treino:", X_train.shape, "| Teste:", X_test.shape)
```

# 5. **Pipeline** de pré-processamento
Essa abordagem permite aplicar de forma sistemática e reprodutível todas as etapas necessárias de tratamento de dados, tanto no conjunto de treino quanto no de teste.


```python
#Separação das colunas numéricas e categóricas
num_cols = [c for c in X_train.columns if str(X_train[c].dtype).startswith(("float","int"))]
cat_cols = [c for c in X_train.columns if c not in num_cols and c != "timestamp"]

#Pipeline para variáveis numéricas
numeric_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

#Pipeline para variáveis categóricas
categorical_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

#Combinação dos pipelines em um pré-processador único
preprocess = ColumnTransformer(transformers=[
    ("num", numeric_pipe, num_cols),
    ("cat", categorical_pipe, cat_cols)
])

#Verificação das colunas
print("num_cols:", num_cols[:5], "...")
print("cat_cols:", cat_cols[:5], "...")
```

Essa etapa garante reprodutibilidade e evita data leakage. Todas as transformações são aprendidas só no treino e aplicadas no teste.

Colunas numéricas (idade, temporada e fifa): valores ausentes imputados pela mediana (mais robusta contra outliers). Escaladas com StandardScaler, para deixar na mesma escala.

Colunas categóricas (posição e lesão): valores ausentes imputados pela moda. Codificadas com OneHotEncoder (handle_unknown="ignore" garante que categorias novas no teste não quebrem o modelo).

```python
preprocess.fit(X_train)

#Recuperar o transformador de categóricas do ColumnTransformer
cat_transformer = preprocess.named_transformers_['cat'].named_steps['onehot']

#Nomes das colunas geradas pelo One-Hot
onehot_columns = cat_transformer.get_feature_names_out(cat_cols)

print("Número de colunas geradas pelo One-Hot Encoding:", len(onehot_columns))
print("Exemplos de colunas:", onehot_columns[:10])
```

```python
#Modelos candidatos
candidates = {
    "LogisticRegression": Pipeline([
        ("pre", preprocess),
        ("model", LogisticRegression(max_iter=200, random_state=SEED))
    ]),
    "RandomForest": Pipeline([
        ("pre", preprocess),
        ("model", RandomForestClassifier(random_state=SEED))
    ])
}

#Treina e avalia cada candidato
for name, model in candidates.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"\nModelo: {name}")
    print("\nAcurácia:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))
```

##6.2 Métricas adicionais

Mesmo com as métricas essenciais demonstradas, é útil acompanhar outras métricas quando há desbalanceamento. As métricas adicionais são:

* Balanced Accuracy, que ajuda a ver se o modelo trata bem as duas classes;

* ROC-AUC, que verifica separação das classes usando probabilidades;

* Matriz de Confusão, que mostra quais classes estão sendo confundidas.

```python
#Binarizar y_test (0 e 1)
lb = LabelBinarizer()
y_test_bin = lb.fit_transform(y_test).ravel()  #.ravel() transforma em 1D

#Probabilidade da classe positiva (classe 1)
y_score = pipe.predict_proba(X_test)[:, 1]  #pega a coluna da classe positiva

#Métricas adicionais
bal_acc = balanced_accuracy_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test_bin, y_score)

print("Balanced Accuracy:", bal_acc)
print("ROC-AUC:", roc_auc)

cm = confusion_matrix(y_test, y_pred)
classes = ['Curta', 'Longa']
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=classes, yticklabels=classes)
plt.ylabel('Classe verdadeira')
plt.xlabel('Classe prevista')
plt.title('Matriz de Confusão - Random Forest')
plt.show()
```

Classe verdadeira: é a informação real que tinha no conjunto de teste (y_test). Representa o que realmente aconteceu, ou seja, a duração da lesão do jogador no dataset (curta ou longa).

Classe prevista: é o resultado que o modelo previu para cada instância do conjunto de teste (y_pred). É a opinião do modelo sobre qual seria a classe correta.

##7.1 Ajuste de hiperparâmetros (tuning)

O dataset não é muito grande. Por essa razão, isso pode limitar a performance dos modelos, especialmente RandomForest, que pode overfitar se não controlar profundidade e número de árvores. Como não dá para aumentar a base, a melhor saída é extrair o máximo de performance com tuning e validação mais robusta.

```python
#Logistic Regression - GridSearchCV
param_grid = {
    "model__C": [0.01, 0.1, 1, 10, 100],
    "model__penalty": ["l2"],
    "model__solver": ["lbfgs", "saga"],
    "model__class_weight": [None, "balanced"]
}

grid_lr = GridSearchCV(
    candidates["LogisticRegression"],
    param_grid=param_grid,
    scoring="f1_weighted",
    cv=5,
    n_jobs=-1
)

grid_lr.fit(X_train, y_train)
print("Melhores parâmetros (Logistic Regression):", grid_lr.best_params_)
print("Melhor f1_weighted:", grid_lr.best_score_)


#RandomForest - RandomizedSearchCV
param_dist = {
    "model__n_estimators": randint(50, 300),
    "model__max_depth": [None, 5, 10, 20],
    "model__min_samples_split": [2, 5, 10],
    "model__min_samples_leaf": [1, 2, 4],
    "model__class_weight": [None, "balanced"]
}

random_rf = RandomizedSearchCV(
    candidates["RandomForest"],
    param_distributions=param_dist,
    n_iter=20,
    scoring="f1_weighted",
    cv=5,
    n_jobs=-1,
    random_state=SEED
)

random_rf.fit(X_train, y_train)
print("Melhores parâmetros (RandomForest):", random_rf.best_params_)
print("Melhor f1_weighted:", random_rf.best_score_)
```
